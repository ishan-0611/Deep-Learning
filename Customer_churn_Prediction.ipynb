### Label Encoding

```python
def print_unique_col_values(df3) :
  for column in df3:
     if df3[column].dtypes == 'object' :
        print(f'{column} : {df3[column].unique()}')
```

```python
print_unique_col_values(df3)
```

```python
df3.replace('No internet service', 'No', inplace = True)
```

```python
df3.replace('No phone service', 'No', inplace = True)
```

```python
print_unique_col_values(df3)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df3[col].replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
for col in df3 :
   print(f'{col} : {df3[col].unique()}')
```

```python
df3['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df3.gender.unique()
```

### One hot encoding for categorical variables

```python
df4 = pd.get_dummies(data=df3, columns=['InternetService','Contract','PaymentMethod'])
df4.columns
```

```python
df4.sample(5)
```

```python
df4.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df4[cols_to_scale] = scaler.fit_transform(df4[cols_to_scale])
```

```python
for col in df4:
    print(f'{col}: {df4[col].unique()}')
```

```python
df4.sample(3)
```

```python
X = df4.drop('Churn', axis = 1)
y = df4['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
len(X_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (26,), activation = 'relu'),
    # keras.layers.Dense(15, activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid'),
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs = 100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:10]
```

```python
y_pred = []
for element in yp:
   if element > 0.5:
      y_pred.append(1)
   else:
      y_pred.append(0)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (8,5))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
